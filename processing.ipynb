```python
import pandas as pd 
from datetime import datetime

df = pd.read_excel("newInput.xlsx")
df = df.dropna(axis=0).reset_index()

def toSeconds(time_str) -> int:
    time_format = "%H:%M:%S"
    # `strptime` is used to parse date object form a string 
    #`time()` is used to access only the time part of the object 
    time = datetime.strptime(time_str, time_format).time()
    #take the hours, minutes and seconds to calculate the total 
    return time.hour * 3600 + time.minute * 60 + time.second
print(toSeconds(df['Departure Time'][1]))
df
```
